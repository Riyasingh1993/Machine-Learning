```python
stopwords = open("stopwords.txt", "r",  encoding="utf8")

arrayofstopwords = []

stopwordsline = stopwords.readlines()

for line in stopwordsline:
    arrayofstopwords.append(line.strip())
```

```python
class ReadFileText:
            
    def __init__(self, text_file_path, permission):
        self.text_file_path = text_file_path
        self.permission = permission
        
        self.perform()
        
    def perform(self):
        self.open_file()
        self.get_words()
        
        
        return self.file_text_words

    def open_file(self):
        self.file_text = open(self.text_file_path, self.permission)
    
    def delete_invalid_chars(self, text):
        for char in "'.!?,<>()-0123456789*\+_!@][#$%:|;" :
            text = text.replace(char, ' ')
            text = text.replace('"', ' ')
            text = text.replace('=', ' ')
            text = text.replace('/', ' ')
            text = text.replace('~', ' ')
            text = text.replace('^', ' ')
            text = text.replace(',', ' ')

        return text
    
    def get_words(self):
        self.file_text_words = []
        
        with self.file_text as f:
            for line in f:
                words = self.delete_invalid_chars(line.decode(errors='ignore')).split()

                for word in words:
                    if word.lower() not in arrayofstopwords:
                        self.file_text_words.append(word.lower())
    
    def closefile(self):
        file_text.close()
```

```python
soc_religion_christian_words = ReadFileText("20_newsgroup/soc.religion.christian.txt/soc.religion.christian.txt", "rb").file_text_words
```

```python
sci_crypt_words = ReadFileText("20_newsgroup/sci.crypt.txt/sci.crypt.txt", "rb").file_text_words
```

```python
def convertarrayinhashwithfrequency(arraywords):
    wordscount = {}
    
    for word in arraywords:
        lowerword = word.lower()
        if lowerword not in arrayofstopwords:
            if lowerword not in wordscount.keys():
                wordscount[lowerword] = 1
            else:
                wordscount[lowerword] += 1
            
    return wordscount

soc_religion_christian_words_count = convertarrayinhashwithfrequency(soc_religion_christian_words)
sci_crypt_words_count = convertarrayinhashwithfrequency(sci_crypt_words)
```

```python
fieldtext1 = list(soc_religion_christian_words_count.keys())
fieldtext2 = list(sci_crypt_words_count.keys())

fieldgeneral = fieldtext1 + fieldtext2

def getuniquevaluesinlist(array):
    arraywithuniquevalues = []
    
    for word in array:
        if word not in arraywithuniquevalues:
            arraywithuniquevalues.append(word)
            
    return arraywithuniquevalues
    
    
uniquefieldnameingeneral = getuniquevaluesinlist(fieldgeneral)
uniquefieldnameingeneral.insert(0, 'label')
```

```python
bagofwords = pd.read_csv("bag_of_words.csv")
bagofwordsphrases = pd.read_csv("bag_of_words_phrases.csv")
tests = pd.read_csv("tests.csv")
```

```python
tests = ["no person is good in himself to go to heaven",
"i am looking for references to algorithms which can be used for password encryption",
"i study religion and read the bible",
"what i meant was that as long as the only advantage of the cryptanalyst is a faster computer",
"then it will be known to the ends of the earth that god rules",
"the fundamental strength of the des and rsa are not nearly so important as what we know about their strength"]

testsresults = [1,2,1,2,1,2]
```

```python
class NaiveBayes:
    
    def __init__(self, bagofwords, bagofwordsphrases, tests, testsresults):
        self.bagofwords = bagofwords
        self.bagofwordsphrases = bagofwordsphrases
        self.testsresults = testsresults
         
        self.text1_words_count = formmatedtext1wordscount
        self.text2_words_count = formmatedtext2wordscount 
        self.text_words_classify()
        
    def perform(self): 
        self.print_results()
        
        index = 0
        checks = 0
        wrongs = 0
        
        for test in tests:                
            prob_text1 = self.p_text1() * self.p_phrase_text1(test)
            prob_text2 = self.p_text2() * self.p_phrase_text2(test)
            
            print("phrase:'" + test + "':")
            
            print("Classification: ", end="")
            if(prob_text1 > prob_text2):
                print("Text 01", end="")
                result = 1
            else:
                print("Text 02", end="")
                result = 2
            
            if(result == testsresults[index]):
                print(" ✓\n")
                checks += 1
            else:
                print(" ✗\n")
                wrongs += 1
                
            
            index += 1
            
        print("\n Accuracy of algorithm: " + str(checks) + "/" + str(len(tests)))
        print('\033[1m'+ str(checks / len(tests) * 100) + "%" + '\033[0m')
                            
    def print_results(self):
        print('\033[1m' + "Natural language processing\n"+ '\033[0m')
        print('\033[1m'+ "Naive Bayes Algorithm\n"+ '\033[0m')
        print('\033[1m'+ "Text 01: soc.religion.christian"+ '\033[0m')
        print('\033[1m'+ "Text 02: sci.crypt\n" + '\033[0m')
        print("13 Sentences for training and 6 Sentences for testing")
        print("Size of vocabulary used: " + str(self.len_vocabulary()) + " words.\n")
        print("Words with the most occurences in Text 01: " + str(self.text_1_total_words))
        print("Words with the most occurences in Text 02: " + str(self.text_2_total_words) + "\n")
        print("Probability that a sentence is from Text 01: "+ str(round(self.p_text1() / self.p_general(), 2)) + " %")
        print("Probability that a sentence is from Text 02: " + str(round(self.p_text2() / self.p_general(), 2)) + " %\n")
        print('\033[1m'+ " Test execution:\n"+ '\033[0m')
        
    def p_phrase_text1(self, phrase_test):
        prob = 1

        for word in phrase_test.split():
            if word not in arrayofstopwords:
                prob = prob * self.p_word_text1(word)
        
        return prob
    
    def p_phrase_text2(self, phrase_test):
        prob = 1
        
        for word in phrase_test.split():
            if word not in arrayofstopwords:
                prob = prob * self.p_word_text2(word)
        
        return prob
    
    def p_word_text1(self, word): 
        return (self.text1_words_count[word] + 1) / (self.text_1_total_words + self.len_vocabulary())
    
    def p_word_text2(self, word):
        return (self.text2_words_count[word] + 1) / (self.text_2_total_words + self.len_vocabulary())
    
    def text_words_classify(self):
        self.text_1_total_words = 0
        self.text_2_total_words = 0
                
        for word in uniquefieldnameingeneral:
            if self.text1_words_count[word] > self.text2_words_count[word]:
                self.text_1_total_words += 1
            else:
                self.text_2_total_words += 1
        
    def len_vocabulary(self):
        return len(bagofwords.values[0]) - 1
    
    def p_text1(self):
        return len(bagofwordsphrases[bagofwordsphrases['label']==0].values)
    
    def p_text2(self):
        return len(bagofwordsphrases[bagofwordsphrases['label']==1].values)
    
    def p_general(self):
        return len(bagofwordsphrases.values)
```

```python
NaiveBayes(bagofwords, bagofwordsphrases, tests, testsresults).perform()
```
